# MLE is a method that finds the most likely value for the parameter based on the data set collected

```python
from  numpy.random import normal
from scipy.stats import norm
import seaborn as sns
import numpy as np
```

**Let us consider the following sample, assuming a Normal Distribution.**

```python
X = [230,250,270]
```

## Normal Distribution - PDF

# Candidates 

Lets consider the following candidates for Maximum Likelihood Estimation(of parameters) for the given sample set.

**N(175,30),** <br>
**N(240,30),** <br>
**N(240,15)**

### Candidate 1

```python
sns.set_style("dark")
sns.displot(normal(175,30,50))
```

### Candidate 2

```python
sns.set_style("dark")
sns.displot(normal(240,30,50))
```

### Candidate 3

```python
sns.set_style("dark")
sns.displot(normal(240,15,50))
```

## Likelihood Function

### Likelihood Function - Candidate 1

```python
first_likelihood = norm.pdf(X[0],175,30) * norm.pdf(X[1],175,30) * norm.pdf(X[2],175,30)
```

### Likelihood Function - Candidate 2

```python
second_likelihood = norm.pdf(X[0],240,30) * norm.pdf(X[1],240,30) * norm.pdf(X[2],240,30)
```

### Likelihood Function - Candidate 3

```python
third_likelihood = norm.pdf(X[0],240,15) * norm.pdf(X[1],240,15) * norm.pdf(X[2],240,15)
```

### Candidate which maximizes likelihood

```python
candidate_selected = np.argmax([first_likelihood,second_likelihood,third_likelihood]) + 1
candidate_selected
```

# Maximum Likelihood Estimate
### mean = 240, std = 15
